```python

df = pd.read_excel('/content/Current-macroeconomic-situation-Tables-Based-on-Six-Month-data-of-2082.83.xlsx', sheet_name = '27.Tourist Arrival ', skiprows = 2)

display(df.head())
```

```python

df_plot = df.drop(columns=['Unnamed: 0'])

# Drop the first row which contains 'Month' and NaN values, as the actual data starts from the second row
df_plot = df_plot.iloc[1:].copy()

# Rename the first column to 'Month' as it contains the month names
df_plot.rename(columns={df_plot.columns[0]: 'Month'}, inplace=True)

# Strip any leading/trailing spaces from the 'Month' column values
df_plot['Month'] = df_plot['Month'].astype(str).str.strip()

# Define a list of valid month names to filter out non-month rows (e.g., 'Total', 'Average')
month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
df_plot = df_plot[df_plot['Month'].isin(month_names)].copy()

# Melt the DataFrame to transform it from wide to long format
# 'Month' will be the identifier variable, and other columns (years) will be unpivoted
df_melted = df_plot.melt(id_vars=['Month'], var_name='Year', value_name='Tourist Arrivals')

# Clean the 'Year' column: remove 'R' or 'P' suffixes (e.g., '2024R' -> '2024')
df_melted['Year'] = df_melted['Year'].astype(str).str.replace(r'[RP]$', '', regex=True)

# Convert 'Year' and 'Tourist Arrivals' columns to numeric, coercing errors to NaN
df_melted['Year'] = pd.to_numeric(df_melted['Year'], errors='coerce')
df_melted['Tourist Arrivals'] = pd.to_numeric(df_melted['Tourist Arrivals'], errors='coerce')

# Drop any rows where 'Year' or 'Tourist Arrivals' values are NaN after conversion
df_melted.dropna(subset=['Year', 'Tourist Arrivals'], inplace=True)

display(df_melted.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15, 8))
sns.lineplot(data=df_melted, x='Year', y='Tourist Arrivals', hue='Month', marker='o')
plt.title('Tourist Arrivals Trend by Month (2005-2026)')
plt.xlabel('Year')
plt.ylabel('Number of Tourist Arrivals')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend(title='Month', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
average_monthly_arrivals = df_melted.groupby('Month')['Tourist Arrivals'].mean().reset_index()

display(average_monthly_arrivals.sort_values(by='Tourist Arrivals', ascending=False))
```

```python
month_to_num = {
    'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr': 4, 'May': 5, 'Jun': 6,
    'July': 7, 'Aug': 8, 'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec': 12
}
df_melted['Month_num'] = df_melted['Month'].map(month_to_num)

correlation = df_melted['Month_num'].corr(df_melted['Tourist Arrivals'])
print(f"Correlation between Month (numerical) and Tourist Arrivals: {correlation:.2f}")
```

The correlation coefficient ranges from -1 to 1. A value close to 1 indicates a strong positive linear relationship, a value close to -1 indicates a strong negative linear relationship, and a value close to 0 indicates a weak or no linear relationship. The calculated correlation of **{correlation:.2f}** suggests a slight positive linear relationship between the month and tourist arrivals. This means that as the month number increases (from January to December), there is a tendency for tourist arrivals to slightly increase, though it's not a strong correlation.

```python
plt.figure(figsize=(12, 7))
sns.barplot(x='Month', y='Tourist Arrivals', hue='Month', data=average_monthly_arrivals.sort_values(by='Tourist Arrivals', ascending=False), palette='viridis', legend=False)
plt.title('Average Tourist Arrivals Per Month (Across All Years)')
plt.xlabel('Month')
plt.ylabel('Average Tourist Arrivals')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

The table above shows the average tourist arrivals for each month, sorted in descending order.

```python
plt.figure(figsize=(12, 7))
sns.scatterplot(x='Month_num', y='Tourist Arrivals', data=df_melted, hue='Month', palette='viridis', alpha=0.7)
plt.title('Tourist Arrivals vs. Numerical Month')
plt.xlabel('Month (Numerical)')
plt.ylabel('Tourist Arrivals')
plt.xticks(ticks=list(month_to_num.values()), labels=list(month_to_num.keys()), rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 8))
sns.boxplot(x='Year', y='Tourist Arrivals', data=df_melted, hue='Year', palette='viridis', legend=False)
plt.title('Distribution of Tourist Arrivals by Year')
plt.xlabel('Year')
plt.ylabel('Tourist Arrivals')
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
