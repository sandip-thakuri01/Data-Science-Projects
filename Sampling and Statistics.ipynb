```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/creditcard.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
data = df['Amount'].dropna().tolist()
```

```python
data.sort()
n = len(data)
```

Finding the quartiles

```python
# Median (Q2)
if n % 2 == 0:
    median = (data[n//2 - 1] + data[n//2]) / 2
else:
    median = data[n//2]

# Lower half and upper half
if n % 2 == 0:
    lower_half = data[:n//2]
    upper_half = data[n//2:]
else:
    lower_half = data[:n//2]
    upper_half = data[n//2+1:]

# Q1
m = len(lower_half)
if m % 2 == 0:
    Q1 = (lower_half[m//2 - 1] + lower_half[m//2]) / 2
else:
    Q1 = lower_half[m//2]

# Q3
m = len(upper_half)
if m % 2 == 0:
    Q3 = (upper_half[m//2 - 1] + upper_half[m//2]) / 2
else:
    Q3 = upper_half[m//2]

print("Q1:", Q1)
print("Median (Q2):", median)
print("Q3:", Q3)
```

Finding the Outliers using IQR method
formula = IQR = Q3-Q1
Lower Bound = Q1 -1.5(IQR)
Upper Bound = Q3 +1.5(IQR)

```python
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = [x for x in data if x < lower_bound or x > upper_bound]

print("IQR:", IQR)
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
print("Number of Outliers:", len(outliers))
```

Findng Means

```python
import math
```

```python
mean = sum(data) / n
print("Mean:", mean)
```

Calculating the Standard Deviation:

```python
variance = sum((x - mean) ** 2 for x in data) / (n - 1)
sd = math.sqrt(variance)

print("Standard Deviation:", sd)
```

```python
# Calculating the Skewness
skewness = (sum((x - mean) ** 3 for x in data) / n) / (sd ** 3)

print("Skewness:", skewness)
```

```python
#caluculating the Kurtosis
kurtosis = (sum((x - mean) ** 4 for x in data) / n) / (sd ** 4)

print("Kurtosis:", kurtosis)
```
